This shows that some stations were never commuting pairs. Majority of station pairs have 1000-2000 trips in total. Only a fewer stations (< 5 stations) serves more than 6000 trips over 4 years. We can target our offer and recommendations to those stations with fewer trips. 

## Question 2. How many subscribers and how do they compare each other with regards to trip?

```python
! bq query --use_legacy_sql=false ' SELECT  subscriber_type, COUNT(*) as total_count FROM `bigquery-public-data.san_francisco.bikeshare_trips` GROUP BY subscriber_type ORDER BY total_count DESC'
```

```python
! bq query --use_legacy_sql=false --max_rows=1000000 --format=csv ' SELECT sub.start_station_name, sub.end_station_name, sub.subscriber_count, cust.customer_count, sub.subscriber_count + cust.customer_count as total_count FROM (SELECT    start_station_name,   end_station_name,   subscriber_type,   COUNT(subscriber_type) as subscriber_count FROM `bigquery-public-data.san_francisco.bikeshare_trips` WHERE subscriber_type = "Subscriber" GROUP BY start_station_name, end_station_name, subscriber_type) as sub INNER JOIN (SELECT    start_station_name,   end_station_name,   COUNT(subscriber_type) as customer_count FROM `bigquery-public-data.san_francisco.bikeshare_trips` WHERE subscriber_type = "Customer" GROUP BY start_station_name, end_station_name, subscriber_type) as cust ON sub.start_station_name = cust.start_station_name AND sub.end_station_name = cust.end_station_name ORDER BY total_count DESC' > subscriber_type.csv
```

```python
subscriber_type = pd.read_csv('subscriber_type.csv')

print("Total count of subscriber types for each station pairs")
display(subscriber_type.head(10))
```

## >> Interpretation on Subscriber types and commuting stations

There are two subscriber types;
1. Subscriber = annual or 30-day member; 
2. Customer = 24-hour or 3-day member

Out of total 983,648 trips, 846,839 trips accounted for subscriber type which is __86%__ of the total trips. As we identified a list of top commuting station pairs in Question 1, we further looked at the subscriber type in each commuting trips. To do so, I created two tables separately and inner joined to split the count of subscriber types for each station pairs. All commuting station pairs have more subscriber_count than customer_count which indicates that majority of bike customers tend to buy monthly passes or annual passes.  

Intringuingly, the top most commuting station __Harry Bridges Plaza (Ferry Building)__ and __Embarcadero at Sansome__ has equally comparable subscriber_count (5483) and customer_count (3667), which accounted for __60%__ and __40%__ respectively. For other station pairs, customer_type significantly dropped. Even the 2nd most commuting station pairs: __San Francsico Caltrain 2 (330 townsend)__ and __Townsend at 7th__ only has 203 customer_type out of 8508 total trips which is only __2%__. This analysis suggests that __Harry Bridges Plaza (Ferry Building)__ and __Embarcadero at Sansome__ would presumably serve as a central location for tourists attraction or for out of states visitors who would probably enjoy one day or 3-day passes as a customer. 

### Recommendation  

GoBike should target stations pairs with high number of "customer_type" for daily passes or weekly passes. Eg., a choice of one-day pass, 3-day pass, or 5-day pass so that a visitor to Bay Area will have a number of choices. 

GoBike should also consider other top commuting stations for commuting fares such as monthly or annual pass with unlimited station choices OR monthly pass with fixed number stations pairs ( < 3 or so) with more discount. Eg, if a customer buys an annual pass for \$149 with unlimited station choices, there's also another available choice of \$99 with limited number of station choices (< 3 or so) where he or she can only rent and return the bikes at certain stations. By doing so will cater to customers need and if in case he or she will need to rent a bike in other stations, they will need to buy a day or 3 day passes for their immediate need. 

## Question 3. Following up on the most commuting trips between stations pairs, how was the utilization of bikes in those commuting stations?

```python
! bq query --use_legacy_sql=false --max_rows=1000000 --format=csv ' SELECT   start_station_name,   no_of_bikes,   total_trips,   ROUND(total_trips/no_of_bikes) as utilization FROM (SELECT start_station_name, COUNT (DISTINCT bike_number) as no_of_bikes, COUNT(*) as total_trips FROM `bigquery-public-data.san_francisco.bikeshare_trips` GROUP BY start_station_name ORDER BY total_trips DESC)' > utilization.csv
```

```python
utilization =pd.read_csv('utilization.csv')

print("List of 5 top stations with highest bikes utilization")
display(utilization.head())

print("List of 5 stations with lowest bikes utilization")
display(utilization.tail())

print("\nTotal number of station for bikes utilization {}".format(len(utilization)))

plt.hist(utilization.utilization)
plt.title("Histogram of Bikes Utilization")
plt.xlabel("Average trips per bike")
plt.ylabel("Number of stations")
plt.show()
```

## >> Interpretation on histogram of bikes utilization rate   
There are more than 40 stations where bikes utilization rate is very low, i.e., each bike has only a few trips per bike. In other stations bikes are heavily utilized with each bike accounting for 140 trips on average per bike. This is where we could target our offers.  

### Recommendations
For stations with very low utilizations such as __Middlefield Light Rail Station, San Jose Governer Center, etc etc__, we can either (1) give an offer such as \$3 for round trips or (2) reduce the number of available bikes and transfer those bikes to highly utilized stations such as __San Francisco Caltrain (Townsend at 4th), San Francisco Caltrain 2 (330 Townsend)__. 

## Question 4: What are the commuting distance between most popular stations?

```python
! bq query --use_legacy_sql=false --max_rows=1000000 --format=csv ' SELECT   start_station_name,   end_station_name,   COUNT(*) as trips,   ROUND(AVG(duration_sec/60), 2) AS avg_duration_min FROM   `bigquery-public-data.san_francisco.bikeshare_trips` GROUP BY   start_station_name, end_station_name ORDER BY   trips DESC' > duration.csv
```

```python
duration = pd.read_csv('duration.csv')

print("List of top commuting stations with their respective average duration")
display(duration.head())

print("\nList of low commuting stations with their respective average duration")
display(duration.tail())

plt.hist(duration.avg_duration_min)
plt.title("Histogram of Average Duration of each trip")
plt.xlabel("Average Duration of each trip in minutes")
plt.ylabel("Number of stations")
plt.show()
```

```python
! bq query --use_legacy_sql=false --format=csv ' SELECT   start_station_name,   end_station_name,   COUNT(*) as trips,   ROUND(AVG(duration_sec/60), 2) AS avg_duration_min FROM   `bigquery-public-data.san_francisco.bikeshare_trips` GROUP BY   start_station_name, end_station_name ORDER BY   trips DESC' > duration_100.csv
```

```python
duration_100 = pd.read_csv('duration_100.csv')

plt.hist(duration_100.avg_duration_min)
plt.title("Histogram of Average Duration of each trip")
plt.xlabel("Average Duration of each trip in minutes")
plt.ylabel("Number of stations")
plt.show()
```

## >> Interpretation on histogram of average duration per trip  

When we analyzed the entire dataset, majority of trip duration was approximately 300 minutes which gave us a range of 300/60 = 5 hours on average. However in order to look at the trip duration at highly commuting stations, we zoomed our analysis only on Top 100 commuting station pairs and investigated their average trip durations. 

The most commuting stations such as __Harry Bridges Plaza (Ferry Building) and Embarcedero Sansome__, accounting for 9150 trips has 19.80 minutes on average for each trip. When we look at the stations that accounted for 1858 trips between __Harry Bridges Plaza (Ferry Building) and Townsend at 7th__, the average trip duration was 16.89 minutes. Overall the commuting duration between each station was less than 30 minutes. It does tell us that trip duration does not affect the total number of trips between stations. Total number of trips between stations would therefore depend on other factors. 

# Question 5. Heatmap of each bike station in San Francisco

## Left Join bikeshare_trips on bikeshare_stations 

I left joined bikeshare_trips data on bikeshare_stations in order to get the latitude and longitude information based on the station_id on bikeshare_stations dataset. 

```python
! bq query --use_legacy_sql=false --max_rows=1000000 --format=csv ' SELECT  t1.start_station_id, t1.start_station_name, t1.end_station_id, t1.end_station_name, t1.duration_sec, t1.start_date, t1.end_date, t1.bike_number, t1.zip_code, t2.station_id, t2.latitude, t2.longitude, t2.dockcount, t2.landmark FROM  `bigquery-public-data.san_francisco.bikeshare_trips` as t1 INNER JOIN `bigquery-public-data.san_francisco.bikeshare_stations` as t2 ON t1.start_station_id = t2.station_id' > trips_lat_long.csv
```

```python
trips_lat_long = pd.read_csv("trips_lat_long.csv")

print("The new table with lat and long information")
trips_lat_long.head()
```

## Retrieving Latitude and Longitude from the combined dataset

```python
! bq query --use_legacy_sql=false --max_rows=1000000 --format=csv ' SELECT   start_station_name,   end_station_name,   latitude,   longitude,   count(*) as total_trips FROM  (SELECT  t1.start_station_id, t1.start_station_name, t1.end_station_id, t1.end_station_name, t1.duration_sec, t1.start_date, t1.end_date, t1.bike_number, t1.zip_code, t2.station_id, t2.latitude, t2.longitude, t2.dockcount, t2.landmark FROM  `bigquery-public-data.san_francisco.bikeshare_trips` as t1 INNER JOIN `bigquery-public-data.san_francisco.bikeshare_stations` as t2 ON t1.start_station_id = t2.station_id)  GROUP BY start_station_name, end_station_name, latitude, longitude ORDER BY total_trips DESC' > trips_map.csv
```

```python
trips_map = pd.read_csv('trips_map.csv')

print("List of highly commuted stations with their respective latitude and longitude information")
display(trips_map.head())

print("\nList of stations with fewer trips with their respective latitude and longitude information")
display(trips_map.tail())
```

## Sanity check after two tables combined and location retrieval

Checking the total_trips after combining two tables, we did not lost any rows or counts because the total trips for each commuting stations were the same as the prior analysis on bikeshare_trips dataset. 



```python
"""Since folium library is not installed in mids image, I'm making this as a comment"""

import folium
from folium.plugins import HeatMap

max_trips = float(trips_map['total_trips'].max())
map_blueprint = folium.Map(location=[37.788000, -122.394203], zoom_start = 14)

map_parameters = HeatMap( list(zip(trips_map['latitude'], trips_map['longitude'], trips_map['total_trips'])),  
                   min_opacity=0.3,
                   max_val=max_trips,
                   radius=15, blur=5, 
                   max_zoom=1, 
                 )

map_blueprint.add_child(map_parameters)

# Saving the heatmap in html
map_blueprint.save("Bike_share_trips.html")

# Display the heatmap
map_blueprint
```
